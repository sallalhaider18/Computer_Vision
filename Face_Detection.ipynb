```python
import cv2
from google.colab.patches import cv2_imshow
#for reading the image
image = cv2.imread('sallalhaider.jpg')
#for converting image into grayscale for better detection
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
#cascade classifier, we usually use for face detection
face_cascade = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
faces = face_cascade.detectMultiScale(gray, 1.7, 3)
print(faces)
for (x,y,w,h) in faces:
    cv2.rectangle(image, (x, y), (x+w, y+h), (255, 0, 0), 2)

    cv2_imshow(image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
