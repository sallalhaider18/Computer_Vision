```python
import cv2
import numpy as np

# Read the image
image = cv2.imread('sallalhaider.jpg')

# Convert image from BGR to HSV color space
hsv_image = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)

# Define lower and upper bounds for the color you want to detect (in HSV)
# For example, here we detect red color
lower_red = np.array([0, 100, 100])
upper_red = np.array([10, 250, 250])


# Create a mask using the specified color range
mask = cv2.inRange(hsv_image, lower_red, upper_red)

# Apply the mask to the original image
color_detected_image = cv2.bitwise_and(image, image, mask=mask)

# Display the original image and the color-detected image
cv2_imshow(image)
cv2_imshow(hsv_image)
cv2_imshow(color_detected_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
